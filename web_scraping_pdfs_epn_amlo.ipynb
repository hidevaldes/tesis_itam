```python
import requests
import pickle
```

```python
urls_epn_amlo=[]
```

```python
with open('urls_epn_amlo.txt') as f:
    urls_epn_amlo = [line.rstrip() for line in f]
```

```python
pages=[]

for i in range(len(urls_epn_amlo)):
    pages.append(requests.get(urls_epn_amlo[i]))
```

```python
page_content=[]
```

```python
for i in range(len(pages)):
    page_content.append(pages[i].content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=[]
```

```python
for i in range(len(page_content)):
    soup.append(BeautifulSoup(page_content[i], 'html.parser'))
```

```python
content_pdfs = []
```

```python
for element in range(len(soup)):
    content_pdfs.append(soup[element].get_text())
```
