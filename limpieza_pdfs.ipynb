```python
import re
import pickle
```

```python
def find_occurrence(occurrence, str):

    inilist = [m.start() for m in re.finditer(r"Informe de Gobierno|informe de gobierno", str)]
    if len(inilist)>= 2:
        return inilist[occurrence - 1]
```

```python
content_pdfs=pickle.load(open(r'C:\Users\HIDEv\OneDrive\Documentos\ArchivosTesis\content_pdfs.pkl','rb'))
```

```python
inicio = []
fin = []

for i in range(len(content_pdfs)):
    inicio.append(find_occurrence(2, content_pdfs[i]))
    fin.append(content_pdfs[i].find('Fuentes:'))
```

```python
auxiliar_pdfs=[]
```

```python
for i in range(len(content_pdfs)):
    auxiliar_pdfs.append(content_pdfs[i][inicio[i]:fin[i]])
```

```python
auxiliar_pdfs[70]=auxiliar_pdfs[70].replace(auxiliar_pdfs[70][auxiliar_pdfs[70].find('Fuente:'):],' ')
```

```python
auxiliar_pdfs[72]=auxiliar_pdfs[72].replace(auxiliar_pdfs[72][auxiliar_pdfs[72].find('Fuente:'):],' ')
```

```python
pdfs_limpios=[]
```

```python
for i in range(len(auxiliar_pdfs)):
    pdfs_limpios.append(auxiliar_pdfs[i].replace('\n',' '))
    pdfs_limpios[i]=pdfs_limpios[i].replace('\t',' ')
```

```python
puntuacion = [',','.',':',';','/','-','–','(',')',"'",'"','!','¡','¿','?','[',']']
```

```python
for i in range(len(pdfs_limpios)):
    for j in range(len(puntuacion)):
        pdfs_limpios[i]=pdfs_limpios[i].replace(puntuacion[j],' ')
```

```python
def remove_numbers(text):
    """
    Remove isolated numbers from text
    :param text: Text to remove numbers for
    :return:
    """
    tokens = text.split()
    tokens = ['' if token.isdigit() else token for token in tokens]
    return ' '.join(tokens)
```

```python
for i in range(len(pdfs_limpios)):
    pdfs_limpios[i]=remove_numbers(pdfs_limpios[i])
```
